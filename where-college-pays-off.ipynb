## Data Cleaning

```python
macro_df.isnull().sum()
```

```python
macro_df = macro_df.dropna()
macro_df.isnull().sum()
```

```python
macro_df.rename(columns={'School Name' : 'School_Name',
                         'School Type' : 'School_Type',
                        'Starting Median Salary_x':'Starting_Median_Salary',
                        'Mid-Career Median Salary_x': 'MidCareer_Median_Salary',
                        'Mid-Career 10th Percentile Salary_x': 'MidCareer_10th_Percentile_Salary',
                        'Mid-Career 25th Percentile Salary_x': 'MidCareer_25th_Percentile_Salary',
                        'Mid-Career 75th Percentile Salary_x': 'MidCareer_75th_Percentile_Salary',
                         'Mid-Career 90th Percentile Salary_x': 'MidCareer_90th_Percentile_Salary'},
                inplace=True)
list(macro_df.columns)
```

```python
macro_df.Starting_Median_Salary = macro_df.Starting_Median_Salary.str.strip('$')
macro_df.MidCareer_Median_Salary = macro_df.MidCareer_Median_Salary.str.strip('$')
macro_df.MidCareer_10th_Percentile_Salary = macro_df.MidCareer_10th_Percentile_Salary.str.strip('$')
macro_df.MidCareer_25th_Percentile_Salary = macro_df.MidCareer_25th_Percentile_Salary.str.strip('$')
macro_df.MidCareer_75th_Percentile_Salary = macro_df.MidCareer_75th_Percentile_Salary.str.strip('$')
macro_df.MidCareer_90th_Percentile_Salary= macro_df.MidCareer_90th_Percentile_Salary.str.strip('$')
```

```python
macro_df
```

```python
macro_df['Starting_Median_Salary'] = macro_df['Starting_Median_Salary'].fillna('').str.replace(',',"").astype(float)
macro_df['MidCareer_Median_Salary'] = macro_df['MidCareer_Median_Salary'].fillna('').str.replace(',',"").astype(float)
macro_df['MidCareer_10th_Percentile_Salary'] = macro_df['MidCareer_10th_Percentile_Salary'].fillna('').str.replace(',',"").astype(float)
macro_df['MidCareer_25th_Percentile_Salary'] = macro_df['MidCareer_25th_Percentile_Salary'].fillna('').str.replace(',',"").astype(float)
macro_df['MidCareer_75th_Percentile_Salary'] = macro_df['MidCareer_75th_Percentile_Salary'].fillna('').str.replace(',',"").astype(float)
macro_df['MidCareer_90th_Percentile_Salary'] = macro_df['MidCareer_90th_Percentile_Salary'].fillna('').str.replace(',',"").astype(float)
```

## Linear Regression

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(macro_dummies, test_size=0.3, random_state=42)
```

```python
feature_cols = [x for x in train.columns if x != 'Starting_Median_Salary']
X_train = train[feature_cols]
y_train = train['Starting_Median_Salary']

X_test  = test[feature_cols]
y_test  = test['Starting_Median_Salary']
```

```python
from sklearn.metrics import mean_squared_error


def rmse(ytrue, ypredicted):
    return np.sqrt(mean_squared_error(ytrue, ypredicted))
```

```python
from sklearn.linear_model import LinearRegression

linearRegression = LinearRegression().fit(X_train, y_train)

linearRegression_rmse = rmse(y_test, linearRegression.predict(X_test))

print(linearRegression_rmse)
```

```python
f = plt.figure(figsize=(6,6))
ax = plt.axes()

ax.plot(y_test, linearRegression.predict(X_test), 
         marker='o', ls='', ms=3.0)

lim = (0, y_test.max())

ax.set(xlabel='Actual Salary', 
       ylabel='Predicted Salary', 
       xlim=lim,
       ylim=lim,
       title='Linear Regression Results');
```

```python
from sklearn.linear_model import RidgeCV

alphas = [0.005, 0.05, 0.1, 0.3, 1, 3, 5, 10, 15, 30, 80]

ridgeCV = RidgeCV(alphas=alphas, 
                  cv=4).fit(X_train, y_train)

ridgeCV_rmse = rmse(y_test, ridgeCV.predict(X_test))

print(ridgeCV.alpha_, ridgeCV_rmse)
```

```python
from sklearn.linear_model import LassoCV

alphas2 = np.array([1e-5, 5e-5, 0.0001, 0.0005])

lassoCV = LassoCV(alphas=alphas2,
                  max_iter=5e4,
                  cv=3).fit(X_train, y_train)

lassoCV_rmse = rmse(y_test, lassoCV.predict(X_test))

print(lassoCV.alpha_, lassoCV_rmse) 
```

```python
from sklearn.linear_model import ElasticNetCV

l1_ratios = np.linspace(0.1, 0.9, 9)

elasticNetCV = ElasticNetCV(alphas=alphas2, 
                            l1_ratio=l1_ratios,
                            max_iter=1e4).fit(X_train, y_train)
elasticNetCV_rmse = rmse(y_test, elasticNetCV.predict(X_test))

print(elasticNetCV.alpha_, elasticNetCV.l1_ratio_, elasticNetCV_rmse)
```

```python
rmse_vals = [linearRegression_rmse, ridgeCV_rmse, lassoCV_rmse, elasticNetCV_rmse]

labels = ['Linear', 'Ridge', 'Lasso', 'ElasticNet']

rmse_df = pd.Series(rmse_vals, index=labels).to_frame()
rmse_df.rename(columns={0: 'RMSE'}, inplace=1)
rmse_df
```

```python
f = plt.figure(figsize=(6,6))
ax = plt.axes()

labels = ['Ridge', 'Lasso', 'ElasticNet']

models = [ridgeCV, lassoCV, elasticNetCV]

for mod, lab in zip(models, labels):
    ax.plot(y_test, mod.predict(X_test), 
             marker='o', ls='', ms=3.0, label=lab)


leg = plt.legend(frameon=True)
leg.get_frame().set_edgecolor('black')
leg.get_frame().set_linewidth(1.0)

ax.set(xlabel='Actual Salary', 
       ylabel='Predicted Salary', 
       title='Linear Regression Results');
```
